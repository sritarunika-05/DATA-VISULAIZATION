```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('/International_Report_Departures.csv')
df1=df.head(50)
df1
```

```python
plt.figure(figsize=(9,9))
plt.scatter(df1["usg_apt_id"],df1["usg_apt"])
plt.xlabel('Usg_apt')
plt.ylabel('Carrier')
plt.title('SCATTER PLOT')
```

```python
plt.figure(figsize=(5,5))
plt.bar(df1["fg_apt"],df1.fg_wac)
plt.xlabel('Fg_apt')
plt.ylabel('Fg_wac')
plt.title('BAR GRAPH')
```

```python
plt.figure(figsize=(5,5))
plt.plot(df1["usg_apt_id"],df1['usg_wac'])
plt.xlabel('usg_apt_id')
plt.ylabel('usg_wac')
plt.title('SIMPLE LINE PLOT')
```

```python
df2=df1.head(5)
plt.figure(figsize=(5,5))
label=df2['fg_apt']
x=df2['fg_wac']
plt.pie(x,labels=label)
plt.title('PIE CHART')
```
